```python
monthly_sales = df.resample(
    'M',
    on='Order Date'
)['Sales'].sum().reset_index()

monthly_sales.head()
```

```python
fig = px.line(

    monthly_sales,

    x='Order Date',

    y='Sales',

    title='📈 Monthly Sales Trend',

    template='plotly_dark'
)

fig.show()
```

```python
region_sales = df.groupby(
    'Region'
)['Sales'].sum().reset_index()

fig = px.bar(

    region_sales,

    x='Region',

    y='Sales',

    color='Sales',

    title='🌍 Region Wise Sales',

    template='plotly_dark'
)

fig.show()
```

```python
category_sales = df.groupby(
    'Category'
)['Sales'].sum().reset_index()

fig = px.pie(

    category_sales,

    names='Category',

    values='Sales',

    title='🛍️ Category Wise Sales',

    template='plotly_dark'
)

fig.show()
```

```python
monthly_sales['Month_Number'] = np.arange(
    len(monthly_sales)
)

monthly_sales['Lag_1'] = monthly_sales[
    'Sales'
].shift(1)

monthly_sales = monthly_sales.dropna()

monthly_sales.head()
```

```python
X = monthly_sales[
    ['Month_Number','Lag_1']
]

y = monthly_sales['Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(

    X,
    y,

    test_size=0.2,

    shuffle=False
)
```

```python
model = LinearRegression()

model.fit(
    X_train,
    y_train
)
```

```python
display(HTML("""

<div style="

background:linear-gradient(
135deg,
#00b09b,
#96c93d
);

padding:30px;

border-radius:25px;

text-align:center;

color:white;

margin-top:30px;
">

<h1>
🤖 Machine Learning Model Trained Successfully
</h1>

</div>

"""))
```

```python
predictions = model.predict(
    X_test
)
```

```python
mae = mean_absolute_error(
    y_test,
    predictions
)

r2 = r2_score(
    y_test,
    predictions
)

print("Mean Absolute Error :", mae)

print("R2 Score :", r2)
```

```python
display(HTML(f"""

<div style="

background:linear-gradient(
135deg,
#1d2671,
#c33764
);

padding:40px;

border-radius:25px;

text-align:center;

color:white;

margin-top:40px;
">

<h1>
📈 Model Performance
</h1>

<h2>
R² Score : {round(r2,2)}
</h2>

<h2>
MAE : {round(mae,2)}
</h2>

</div>

"""))

```

```python
comparison_df = pd.DataFrame({

    'Actual': y_test,

    'Predicted': predictions

})

fig = px.line(

    comparison_df,

    title='📊 Actual vs Predicted Sales',

    template='plotly_dark'
)

fig.show()
```

```python
future_months = np.arange(

    len(monthly_sales),

    len(monthly_sales)+12

)

last_sales = monthly_sales[
    'Sales'
].iloc[-1]

future_input = pd.DataFrame({

    'Month_Number': future_months,

    'Lag_1': [last_sales]*12
})

future_predictions = model.predict(
    future_input
)

future_predictions
```

```python
future_dates = pd.date_range(

    start=monthly_sales[
        'Order Date'
    ].max(),

    periods=12,

    freq='ME'
)

future_dates
```

```python
forecast_df = pd.DataFrame({

    'Date': future_dates,

    'Forecast Sales': future_predictions

})

forecast_df
```
